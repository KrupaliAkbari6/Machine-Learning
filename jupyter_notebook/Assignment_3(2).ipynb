```python
# import pandas as pd
# import numpy as np
# from sklearn import linear_model,model_selection,preprocessing
# import matplotlib.pyplot as plt


# data=pd.read_csv("Weather.csv")
# ndf=data[["Snowfall"]]
# ndf.dropna(axis=0,inplace=True)# 0=row ,1=column
# # ndf.isnull().sum()
# # print(data)

# # # Load The Data


# #split the daset
# 
# y=np.array(ndf).reshape(-1,1)
# print(x)x=np.array(data.MinTemp).reshape(-1,1)
# print(y)


# #load the Model
# lr=linear_model.LinearRegression()

# #Trait the model:-fit
# lr.fit(x,y)

# #Test the model:
# xtest=x
# ytest_result=lr.predict(xtest)
# print(xtest)
# print(y)
# print(ytest_result)

# # visualize the actual value v/s predicted value
# plt.scatter(x,y,color='red')
# plt.plot(x,ytest_result,color="blue")
# plt.xlabel("XTEST")
# plt.ylabel("YTEST")
# plt.show()

# data=pd.read_csv("Weather.csv")

# #replace
# data.replace("#VALUE!",np.nan,inplace=True)

# data.isnull().sum()
# ndf=data[['MinTemp','Snowfall']]

# print(ndf.isnull().sum())

# x=np.array(data.MinTemp).reshape(-1,1)
# y=np.array(pd.to_numeric(ndf.Snowfall)).reshape(-1,1)
# print(ndf.dtypes)

# xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y,test_size=0.1)

# #load model
# lr=linear_model.LinearRegression()

# #train the Model:fit()
# lr.fit(xtrain,ytrain)
# ytest_predicted=lr.predict(xtest)

# print(ytest)
# print(ytest_predicted)


# plt.scatter(xtrain,ytrain,color='red')
# plt.scatter(xtest,ytest_predicted,color="blue")
# plt.xlabel("XTEST")
# plt.ylabel("YTEST")
# plt.show()



import pandas as pd
import numpy as np
from sklearn import linear_model, model_selection
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("Weather.csv")

# Replace any invalid values (e.g., "#VALUE!") with NaN
data.replace("#VALUE!", np.nan, inplace=True)

# Check for null values
print(data.isnull().sum())

# Drop rows with null values in relevant columns
data.dropna(subset=['MinTemp', 'Snowfall'], inplace=True)

# Prepare the features (X) and target (y)
x = np.array(data['MinTemp']).reshape(-1, 1)
y = np.array(pd.to_numeric(data['Snowfall'], errors='coerce')).reshape(-1, 1)

# Check data types
print("Data types:")
print(data.dtypes)

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.1, random_state=42)

# Load the linear regression model
lr = linear_model.LinearRegression()

# Train the model
lr.fit(x_train, y_train)

# Predict using the test set
y_test_predicted = lr.predict(x_test)

# Print the actual and predicted values for comparison
print("Actual values (y_test):")
print(y_test.flatten())
print("Predicted values (y_test_predicted):")
print(y_test_predicted.flatten())

# Visualize the training and predicted values
plt.scatter(x_train, y_train, color='red', label='Training data')
plt.scatter(x_test, y_test_predicted, color='blue', label='Predicted data')
plt.xlabel("Min Temperature")
plt.ylabel("Snowfall")
plt.legend()
plt.title("Linear Regression: Min Temperature vs Snowfall")
plt.show()


```

# Data with Insurance on linear regression eith Error

```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model,model_selection,preprocessing
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error

data=pd.read_csv('insurance.csv')
# print (data.to_string())

#indentify x and y
x=np.array(data.sex).reshape(-1,1)
y=np.array(data.expenses).reshape(-1,1)

#data preprocessing
labeled_data=preprocessing.LabelEncoder()
op=np.array(labeled_data.fit_transform(data.sex)).reshape(-1,1)
print(op)

#split the database in training and testing  
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(op,y,test_size=0.3)

#load the Model
lr=linear_model.LinearRegression()

#Trait the model:-fit
lr.fit(xtrain,ytrain)

#Test the model:
ytest_result=lr.predict(xtest)
# print(ytest)
# print(ytest_result)

# visualize the actual value v/s predicted value
plt.scatter(xtrain,ytrain,color='red')
plt.plot(xtest,ytest_result,color="blue")
plt.xlabel("sex")
plt.ylabel("expenses")
plt.show()

#error
print("mean absolute error:" ,mean_absolute_error(ytest,ytest_result))
print("mean square error:" ,mean_squared_error(ytest,ytest_result))
print("mean absolute percentage error:" ,mean_absolute_percentage_error(ytest,ytest_result))

```
