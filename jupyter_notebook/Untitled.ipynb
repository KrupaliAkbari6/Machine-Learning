# layout (subplot)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,10))
img=cv2.imread("men.webp")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(1,3,1)
plt.imshow(new_img)
plt.axis('Off')

k1=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
fitered_img=cv2.filter2D(new_img,-1,k1)
plt.subplot(1,3,2)
plt.imshow(fitered_img)

blur_img=cv2.blur(new_img,(10,10))
plt.subplot(1,3,3)
plt.imshow(blur_img)
```

# Box Blur, Gaussian Blur and Median Blur


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,10))
# img=cv2.imread("nose.jpeg")
img=cv2.imread("men.webp")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
box_blurred=cv2.blur(new_img,(10,10))
gaussian_blurred=cv2.GaussianBlur(new_img,(5,5),0)
median_blurred=cv2.medianBlur(new_img,3)


plt.subplot(1,4,1),plt.imshow(new_img),plt.title("Original Image"),plt.axis('Off')
plt.subplot(1,4,2),plt.imshow(box_blurred),plt.title("Blur Image"),plt.axis('Off')
plt.subplot(1,4,3),plt.imshow(gaussian_blurred),plt.title("Gaussian Blur Image"),plt.axis('Off')
plt.subplot(1,4,4),plt.imshow(median_blurred),plt.title("Median Blur Image"),plt.axis('Off')

```

# Sharpening (Border)


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("nose.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

k1=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
k2=np.array([[1,1,1],[1,-7,1],[1,1,1]])

img1=cv2.filter2D(new_img,-1,k1)
img2=cv2.filter2D(new_img,-1,k2)

plt.subplot(1,2,1),plt.imshow(img1),plt.axis('Off')
plt.subplot(1,2,2),plt.imshow(img2),plt.axis('Off')
```

# Embossing (Shadow)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("nose.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

k1=np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
k2=np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
k3=np.array([[0,0,1],[0,0,0],[-1,0,0]])
k4=np.array([[1,0,0],[0,0,0],[0,0,-1]])

img1=cv2.filter2D(new_img,-1,k1)+128
img2=cv2.filter2D(new_img,-1,k2)+128
img3=cv2.filter2D(new_img,-1,k3)+128
img4=cv2.filter2D(new_img,-1,k4)+128

plt.subplot(1,4,1),plt.imshow(img1),plt.title("South East"),plt.axis('Off')
plt.subplot(1,4,2),plt.imshow(img2),plt.title("South West"),plt.axis('Off')
plt.subplot(1,4,3),plt.imshow(img3),plt.title("North East"),plt.axis('Off')
plt.subplot(1,4,4),plt.imshow(img4),plt.title("North West"),plt.axis('Off')
```

# Sobel (Edges Detection)

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

img=cv2.imread("nose.jpeg")
# new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(1,3,1)
plt.imshow(new_img,cmap="gray")
plt.title("Original Image")
plt.axis("off")

sobelx=cv2.Sobel(new_img,cv2.CV_64F,1,0,5)
plt.subplot(1,3,2)
plt.imshow(sobelx,cmap="gray")
plt.title("Horizontal Edges ")
plt.axis("off")

sobely=cv2.Sobel(new_img,cv2.CV_64F,0,1,5)
plt.subplot(1,3,3)
plt.imshow(sobely,cmap="gray")
plt.title("vertical Edges ")
plt.axis("off")
```

# Laplacian and Canny

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

img=cv2.imread("nose.jpeg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(1,3,1)
plt.imshow(new_img,cmap="gray")
plt.title("Original Image")
plt.axis("off")

laplacian_img=cv2.Laplacian(new_img,cv2.CV_64F)
plt.subplot(1,3,2)
plt.imshow(laplacian_img)
plt.title("Laplacian Edges ")
plt.axis("off")

canny_img=cv2.Canny(new_img,50,150)
plt.subplot(1,3,3)
plt.imshow(canny_img,cmap="gray")
plt.title("Canny Edges ")
plt.axis("off")
```
