# Cluster

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import MeanShift, estimate_bandwidth
```

# Clustering on Book dataset

```python
#load the dataset
data=pd.read_csv("Book.csv")
x=data.iloc[:,0]
y=data.iloc[:,1]

#plot the dataset to identify the clusters
plt.figure()
plt.scatter(x,y,marker="o",facecolors="none",edgecolors="k",s=30)
plt.title("Input data")
plt.xticks(())
plt.yticks(())

#Load the model
kmeans=KMeans(init="k-means++",n_clusters=4)

#Train the model
kmeans.fit(data)

#Use the model to predict values
results=kmeans.predict(data)
print(results)

#Find the centroids
centroids = kmeans.cluster_centers_
print(centroids)

#plot the clusters along with centroids
plt.figure()
plt.scatter(x,y,marker="o",facecolors="none",edgecolors="k",s=30)
plt.scatter(centroids[:,0],centroids[:,1],marker="o",s=200,color="r")
plt.title("Centroids on boundaries obtaines using KMeans")
```

# Clustering on clustering dataset

```python
#load the dataset
data=pd.read_csv("clustering.csv")

x=data["ApplicantIncome"]
y=data["LoanAmount"]

data=list(zip(x,y))

#plot the dataset to identify the clusters
plt.figure()
plt.scatter(x,y,marker="o",facecolors="none",edgecolors="k",s=30)
plt.title("Input data")
plt.xticks(())
plt.yticks(())

#Load the model
kmeans=KMeans(init="k-means++",n_clusters=4)

#Train the model
kmeans.fit(data)

#Use the model to predict values
results=kmeans.predict(data)
print(results)

#Find the centroids
centroids = kmeans.cluster_centers_
print(centroids)



#plot the clusters along with centroids
plt.figure()
plt.scatter(x,y,c=results,marker="o",facecolors="none",edgecolors="k",s=30)
plt.scatter(centroids[:,0],centroids[:,1],marker="o",s=200,color="r")
plt.title("Centroids on boundaries obtaines using KMeans")
```

# MeanShift clustering with 2017 dataset

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

#load the dataset
data = pd.read_csv("2017.csv")

#preprocess
data = data.iloc[:,1:]

#claculate the bandwidth
bandwidth = estimate_bandwidth(data)

#load the model
meanshift = MeanShift(bandwidth=bandwidth)

#train the model
result = meanshift.fit(data)

#show the results
result.predict(data)
```
